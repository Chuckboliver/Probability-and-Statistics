# **Outlier Detection**

## **ใช้วิธี IQR(Interquartile Range) Outlier Detection Method**
>* IQR คือ ผลต่างระหว่าง Percentile ที่ 75 และ Percentile ที่ 25 เราสามารถใช้ค่า IQR ในการตรวจจับ Outlier ได้

## ***ฟังก์ชั่นในการหา Percentile ที่ n ของ pandas.DataFrame***
 **Parameters :**
> * **df (pandas.DataFrame):** DataFrame ของ Pandas ที่ต้องการนำมาหา Percentile
* **percentile_rank (float):** อันดับของ Percentile ที่ต้องการหา เช่น Percentile ที่ 75
* **column_name (str):** ชื่อของคอลัมน์ที่ต้องการหา Percentile

**Returns :**
> * **Tuple[str, float]:** ผลลัพธ์จากการหา Percentile ของ DataFrame ค่าแรกของ tuple คือ String ชื่อของ Pokemon ส่วนค่าที่สองของ tuple คือ Integer คือค่าของ Percentile ที่ต้องการหาในคอลัมน์ที่กำหนด



```
def get_percentile(df: pd.DataFrame, percentile_rank: int, column_name: str) -> Tuple[str, float]:
  df_1 = df.sort_values(by=column_name).reset_index()

  index = (len(df_1.index)-1) * percentile_rank / 100.0
  index = int(index)

  return (df_1.at[index, "Name"], df_1.at[index, column_name])
```

## ***ฟังก์ชั่นในการหาค่า IQR***
**Parameters :**
> * **df (pandas.DataFrame):** DataFrame ของ Pandas ที่ต้องการนำมาหาค่าIQR
* **column_name (str):** ชื่อของคอลัมน์ที่ต้องการหาค่า IQR

**Returns :**
> * **float :** ค่าของ IQR

```
def interquartile_range(df: pd.DataFrame, column_name: str) -> float:
  c75, p75 = get_percentile(df, 75, column_name)
  c25, p25 = get_percentile(df, 25, column_name)
  return p75 - p25
```

## ***ฟังก์ชั่นในการหา Outlier***
**Parameters :**
> * **df (pandas.DataFrame):** DataFrame ของ Pandas ที่ต้องการนำมาหา Outliers
* **column_name (str):** ชื่อของคอลัมน์ที่ต้องการหา Outliers

**Returns :**
> * **pandas.DataFrame :** DataFrame ของ Outlier

```
def get_outliers(df: pd.DataFrame, column_name: str) -> pd.DataFrame:
  c25, p25 = get_percentile(df, 25, column_name)
  c75, p75 = get_percentile(df, 75, column_name)
  iqr = interquartile_range(df, column_name)
  lower_bound = p25 - 1.5*iqr
  upper_bound = p75 + 1.5*iqr
  mask = (df[column_name] < lower_bound) | (df[column_name] > upper_bound)
  return df[mask]
```
