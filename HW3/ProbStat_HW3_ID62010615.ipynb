# **ตารางข้อมูล**
---
* ข้อมูลขนาด 721 แถว

## **Columns ที่สนใจ :**
1. **Name** (ชื่อของ Pokemon)
2. **Height_m** (ส่วนสูงของ Pokemon ในหน่วยเมตร)
3. **Weight_kg** (น้ำหนักของ Pokemon ในหน่วยกิโลกรัม)
4. **Type_1** (ชนิดของ Pokemon)
5. **Body_Style** (รูปร่างร่างกายของ Pokemon)


```
df = pd.read_csv("/content/drive/MyDrive/Data sets/pokemon_alopez247.csv")
df = df[["Name", "Height_m", "Weight_kg", "Type_1",  "Body_Style"]]
df
```

# **Probability Density Function**
---
* Plot graph **Probability Density Function** 1 กราฟต่อข้อมูล 1 คอลัมน์

##***Height(m)***
> ข้อมูลจากคอลัมน์ **Height_m**

```
ax = sns.kdeplot(df["Height_m"], color="red")
ax.set_xlabel("Height(m)")
ax.set_title("Probability Density Function of Pokemon's height")
plt.show()
```

## ***Weight(kg)***
> ข้อมูลจากคอลัมน์ **Weight_kg**

```
ax = sns.kdeplot(df["Weight_kg"], color="green")
ax.set_xlabel("Weight(kg)")
ax.set_title("Probability Density Function of Pokemon's weight")
plt.show()
```

# **Cumulative Distribution Function**
---
* Plot graph **Cumulative Distribution Function** 1 กราฟต่อข้อมูล 1 คอลัมน์

##***Height(m)***

```
ax = sns.kdeplot(df["Height_m"], color="red", cumulative=True)
ax.set_xlabel("Height(m)")
ax.set_title("Cumulative Distribution Function of Pokemon's height")
plt.show()
```

## ***Weight(kg)***

```
ax = sns.kdeplot(df["Weight_kg"], color="green", cumulative=True)
ax.set_xlabel("Weight(kg)")
ax.set_title("Cumulative Distribution Function of Pokemon's weight")
plt.show()
```

# **บทวิเคราะห์ข้อมูลจากกราฟ**
---

จากกราฟ **Probability Density** เป็นกราฟที่ใช้ข้อมูลจากคอลัมน์ความสูง **(Height_m)** และคอลัมน์น้ำหนัก**(Weight_kg)** ของ **Pokemon ทั้งหมด 721 ตัว** จะเห็นได้ว่าที่กราฟความสูงจะมีความหนาแน่นมากในช่วงความสูง 0.1 ถึง 2.3 เมตร และน้ำหนักในช่วง 0.1 ถึง 136 เมตร ทำให้วิเคราะห์ได้ว่า Pokemon ส่วนใหญ่จะมี**ความสูง 0.1 ถึง 2.3 เมตร** และ**น้ำหนักในช่วง 0.1 ถึง 136 เมตร** 

จากกราฟ **Cumulative Distribution** เป็นกราฟที่ใช้ข้อมูลจากคอลัมน์ความสูง **(Height_m)** และคอลัมน์น้ำหนัก**(Weight_kg)** ของ**Pokemon ทั้งหมด 721 ตัว** จะเห็นได้ว่าที่กราฟความสูงจะมีอัตราการเพิ่มของความหนาแน่นที่สูงในช่วง 0.1 ถึง 2.3 เมตร และที่กราฟน้ำหนักก็จะมีอัตราการเพิ่มของความหนาแน่นที่สูงในช่วง 0.1 ถึง 136 กิโลกรัม ซึ่งเป็นไปตามกราฟ **Probability Density**

สามารถสรุปได้ว่าการใช้กราฟ **Probability Density** และ **Cumulative Distribution** สามารถนำมาวิเคราะห์ข้อมูลของประชากร Pokemon ได้ว่ามีความหนาแน่นในช่วงไหนและสามารถนำไปศึกษาต่อได้
